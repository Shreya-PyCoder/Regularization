```python
import pandas as pd
```

```python
A=pd.read_csv("F:/data science workshop/DATA-SET/data/iris.csv")
```

```python
A.head()
```

```python
A=A.drop(labels="Unnamed: 0",axis=1)
```

```python
import seaborn as sb
```

```python
import matplotlib.pyplot as plt
```

```python
sb.heatmap(A.corr())
```

```python
for i in A.columns:
    if(A[i].dtypes=="object"):
        sb.boxplot(A[i],A["Petal.Length"])
        plt.show()
    else:
        sb.scatterplot(A[i],A["Petal.Length"])
        plt.show()
```

```python
X=A[["Petal.Width","Sepal.Length","Sepal.Width"]]
Y=A[["Petal.Length"]]
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
model=lr.fit(xtrain,ytrain)
```

```python
pred=model.predict(xtest)
```

```python
pred
```

```python
ytest
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
print(mean_absolute_error(ytest,pred))
```

```python
print(mean_squared_error(ytest,pred))
```

```python
from sklearn.metrics import explained_variance_score
```

```python
ev=explained_variance_score(ytest,pred)
```

```python
ev
```

# RIDGE REGRESSION

```python
q=1
tpr=[]
for i in range(1,1000,1):
    tpr.append(q+0.01)
    q=q+0.01
```

```python
tpr
```

```python
from sklearn.linear_model import Ridge
```

```python
rr=Ridge()
tp={"alpha":tpr}
```

```python
from sklearn.model_selection import GridSearchCV
cv=GridSearchCV(rr,tp,scoring="explained_variance",cv=4)
cvmodel=cv.fit(xtrain,ytrain)
```

```python
cvmodel.best_params_
```

```python
rr=Ridge(alpha=1.01)
rrmodel=rr.fit(xtrain,ytrain)
rrpred=rrmodel.predict(xtest)
```

```python
print(mean_absolute_error(ytest,rrpred))
```
